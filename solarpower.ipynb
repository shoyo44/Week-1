## ML model for solar prediction


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
df.columns
```

```python
X=df.drop('generated_power_kw',axis=1)
y=df['generated_power_kw']
```

```python
X.shape
```

```python
y.shape
```

```python
# testing the model
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
#Scaling
scale=StandardScaler()
X_train_scaling=scale.fit_transform(X_train)
X_test_scaling=scale.transform(X_test)
```

```python
model=LinearRegression()
model.fit(X_train_scaling,y_train)
```

```python
y_hat=model.predict(X_train_scaling)
y_actual=mean_squared_error(y_train,y_hat)
print(f"The mean square error of trained set is: {y_actual}")
```

```python
y_pred=model.predict(X_test_scaling)
y_predict=mean_squared_error(y_test,y_pred)
print(f"The mean square error of test set is: {y_predict}")
```
