```python
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=30,kde=True)
plt.title('Distribution of power')
plt.xlabel('Generated power')
plt.ylabel('Frequency')
plt.show()
```

```python
df.columns
```

```python
##scatter plot
plt.figure(figsize=(15,30))
for i, col in enumerate(df.columns):
    if(col!='generated_power_kw'):
       plt.subplot(7,3,i+1)
       plt.scatter(df[col],df['generated_power_kw'])
       plt.title(f"{col} vs generated power(KW)")
       plt.xlabel(col)
       plt.ylabel("Genearted power (KW)")
plt.tight_layout()
plt.show()
```

```python
##correlation
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),cmap='viridis',annot=True,fmt='0.2f')
plt.show()
```

# Outliers Detection

```python
## outlier detection
plt.figure(figsize=(15,30))
for i, col in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    sns.boxplot(df[col])
plt.tight_layout()
plt.show()
    
    
```
